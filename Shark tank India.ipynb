# Machine Learning (predict Deal/No Deal)

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
x = cleaned_df[["pitcher_ask_amount", "ask_equity", "ask_valuation"]]
y = cleaned_df["deal"]
```

```python
clf = RandomForestClassifier(max_depth=8, random_state=0)
```

```python
reg = clf.fit(x,y)
```

```python
reg.score(x,y)
```

```python
clf.predict_proba([[10, 2, 20000]])
```
